## Section 1 — Introduction to Pandas: Check Your Understanding

> Reference: [Pandas indexing documentation](https://pandas.pydata.org/docs/user_guide/indexing.html)

### Q1: What is the difference between `.loc[]` and `.iloc[]`? Give one example of each.

**Answer:**

- **`.loc[]`** uses **labels** — it accepts row index labels and column *names*. When the index is the default integer index, `.loc[0]` selects the row with *label* 0 (which may not be the first row if the DataFrame has been filtered and re-indexed).
- **`.iloc[]`** uses **integer positions** — it always counts from 0, regardless of the index labels.

```python
df.loc[0, 'sex']        # row with label 0, column named 'sex'
df.iloc[0, 0]           # row at position 0, column at position 0
df.loc[10:15, ['sex', 'length']]   # label-based slice
df.iloc[10:15, [0, 1]]             # position-based slice
```

> Key gotcha: `.loc` slices are **inclusive on both ends**; `.iloc` slices follow Python convention and are **exclusive on the right end**.

### Q2: You run `df['age'].mean()` and get `NaN`. What is the most likely cause?

**Answer:**

The column contains one or more **`NaN` values**. By default, most NumPy and Pandas aggregation functions return `NaN` if any input is `NaN`. To get a result that ignores missing values, use:

```python
df['age'].mean()              # returns NaN if any value is NaN
df['age'].mean(skipna=True)   # default — but only works if column dtype is numeric
df['age'].dropna().mean()     # explicitly drop NaNs first
```

A secondary cause could be that the column's dtype is `object` (string), in which case `.mean()` is not defined. Check with `df['age'].dtype`.

> Reference: [Pandas missing data guide](https://pandas.pydata.org/docs/user_guide/missing_data.html)

### Q3: What does `.describe()` NOT show, and why?

**Answer:**

By default, `.describe()` only summarises **numeric columns**. Non-numeric columns (like `sex` in the abalone dataset, which has dtype `object`) are excluded because statistics like mean and standard deviation are meaningless for categorical data.

To include non-numeric columns, pass `include='all'`:

```python
df.describe(include='all')   # includes object/category columns
df.describe(include='O')     # only object columns
```

For non-numeric columns, `.describe()` instead shows `count`, `unique`, `top` (most frequent value), and `freq` (frequency of the top value).

> Reference: [DataFrame.describe()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html)

### Q4: You want to find all rows where `weight_whole > 1.0` AND `sex == 'M'`. Write the boolean mask filter.

**Answer:**

````python
import pandas as pd
df = pd.read_csv('data/abalone.csv')

# Boolean mask — each condition must be wrapped in parentheses
# Use & for AND, | for OR
result = df[(df['weight_whole'] > 1.0) & (df['sex'] == 'M')]

# Alternatively, using .query() — often more readable
result_q = df.query('weight_whole > 1.0 and sex == "M"')

print(result.shape)
````

> Reference: [Boolean indexing](https://pandas.pydata.org/docs/user_guide/indexing.html#boolean-indexing) | [.query()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html)

### Q5: What is the difference between `df.drop('col', axis=1)` and `df.drop('col', axis=1, inplace=True)`?

**Answer:**

- `df.drop('col', axis=1)` — returns a **new DataFrame** with the column removed. The original `df` is unchanged.
- `df.drop('col', axis=1, inplace=True)` — modifies `df` **directly** and returns `None`.

```python
new_df = df.drop('col', axis=1)    # df still has 'col'; new_df does not
df.drop('col', axis=1, inplace=True)  # df no longer has 'col'
```

Under Pandas 3.0's **Copy-on-Write** model, `inplace=True` will trigger a copy anyway in many cases. Explicit reassignment (`df = df.drop(...)`) is becoming the preferred style.

> Reference: [DataFrame.drop()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html) | [Copy-on-Write guide](https://pandas.pydata.org/docs/user_guide/copy_on_write.html)

## Section 3 — Introduction to Visualization: Check Your Understanding + Challenges

### Check Your Understanding Q1: What does a histogram tell you that `.describe()` does not?

**Answer:**

`.describe()` gives point-statistics (mean, std, quartiles), but it gives no information about **shape**. A histogram shows:

- Whether the distribution is **symmetric**, **skewed left**, or **skewed right**
- Whether there are **multiple peaks** (bimodal or multimodal distributions)
- Where **clusters** or **gaps** exist in the data
- The **overall spread and density** of values across the range

Two distributions can have identical mean and standard deviation but look completely different when plotted.

> Reference: [Histogram explanation — Wikipedia](https://en.wikipedia.org/wiki/Histogram)

### Check Your Understanding Q2: You want to see if higher alcohol content is associated with higher quality. Which chart type?

**Answer:**

A **bar chart** (groupby + mean) is clearest here, because `quality` is a discrete variable. It shows the average alcohol content at each quality score in a single, easy-to-read view.

A **scatter plot** also works — it shows all individual data points and reveals the variation within each quality score. The scatter is noisier but shows more information.

A **box plot** grouped by quality is another good option — it shows the spread and outliers per quality level, not just the average.

> Reference: [Choosing chart types — Pandas docs](https://pandas.pydata.org/docs/user_guide/visualization.html)

### Check Your Understanding Q3: A box plot shows a very long upper whisker. What does that suggest?

**Answer:**

A long upper whisker means the **upper tail of the distribution is stretched out** — there are values significantly higher than the 75th percentile (Q3), but still within the non-outlier range (typically 1.5 × IQR from Q3). If any dots appear above the whisker, those are **outliers** — individual values that fall beyond that range.

This pattern is typical of **right-skewed distributions** (e.g. income, price, count data) where most values are clustered at the low end but a few are very high.

> Reference: [Box plot interpretation](https://en.wikipedia.org/wiki/Box_plot)

### Check Your Understanding Q4: Why is it a problem to plot all columns together when scales differ?

**Answer:**

When columns have very different scales (e.g. `total sulfur dioxide` ranges from 6 to 289, while `density` ranges from 0.99 to 1.004), all columns share the **same y-axis** in a combined plot. Columns with small ranges get **compressed flat**, making it impossible to see their distribution. Only the columns with large absolute values remain readable.

Solutions:
- Select columns with **similar scales** for a single plot
- Use `subplots=True` to give each column its own y-axis

```python
df.plot(kind='box', subplots=True, layout=(3, 4), figsize=(14, 8))
plt.tight_layout()
plt.show()
```

## Section 6 — Introduction to NumPy: Check Your Understanding

> Source: `2_Numpy/01_Introduction_to_Numpy.ipynb`  
> Reference: [NumPy user guide](https://numpy.org/doc/stable/user/index.html) | [NumPy quickstart](https://numpy.org/doc/stable/user/quickstart.html)

### Q1: What are the two main reasons NumPy arrays are faster than Python lists?

**Answer:**

1. **Contiguous memory**: All elements are stored in a single, unbroken block of memory. Python lists store *pointers* to objects scattered across memory, causing overhead for every element access.

2. **Homogeneous types**: Every element in a NumPy array has the same data type (e.g. `float64`). Python lists can hold mixed types, so Python must type-check every element before each operation. NumPy skips this entirely.

Together these properties enable **vectorisation** — the CPU can execute operations on multiple values simultaneously using SIMD (Single Instruction, Multiple Data) instructions.

> Reference: [Why NumPy is fast — NumPy docs](https://numpy.org/doc/stable/user/whatisnumpy.html)

### Q2: `arr = np.array([1, 2, 'three', 4])`. What is `arr.dtype`? Why?

**Answer:**

`arr.dtype` will be a **Unicode string type** (shown as something like `<U21`). Because NumPy arrays must be homogeneous, when you mix integers and strings, NumPy **upcasts** all values to the most general type — in this case, Unicode string. The integers `1`, `2`, and `4` are silently converted to strings `'1'`, `'2'`, `'4'`.

```python
import numpy as np
arr = np.array([1, 2, 'three', 4])
print(arr.dtype)    # e.g. '<U21'
print(type(arr[0])) # <class 'numpy.str_'>
```

> Reference: [NumPy dtype documentation](https://numpy.org/doc/stable/reference/arrays.dtypes.html)

### Q3: You have a 3×4 NumPy array. What does `arr.sum(axis=1)` return?

**Answer:**

**3 values** — one sum per row. `axis=1` means "collapse across the columns", so each row is reduced to a single value. For a 3×4 array, the result has shape `(3,)`.

```
axis=0 → sum down the rows  → result shape: (4,)  — one value per column
axis=1 → sum across columns → result shape: (3,)  — one value per row
no axis → grand total       → result shape: ()    — single scalar
```

> Reference: [NumPy sum with axis](https://numpy.org/doc/stable/reference/generated/numpy.sum.html)

### Q4: What is the difference between `arr.argmax()` and `arr.max()`?

**Answer:**

- **`arr.max()`** returns the **value** of the largest element in the array.
- **`arr.argmax()`** returns the **index** (position) of the largest element — i.e. *where* the maximum is, not what it is.

```python
arr = np.array([3, 7, 2, 9, 1])
arr.max()     # → 9
arr.argmax()  # → 3  (the maximum is at index 3)
```

> Reference: [numpy.argmax()](https://numpy.org/doc/stable/reference/generated/numpy.argmax.html)

### Q5: You want to find all positions where the value is greater than 100. Which function?

**Answer:**

**`np.where(arr > 100)`** — it returns a tuple of arrays containing the **indices** of all elements that satisfy the condition.

```python
arr = np.array([50, 120, 80, 200, 95])
np.where(arr > 100)  # → (array([1, 3]),)

# To get the values at those indices:
arr[np.where(arr > 100)]  # → array([120, 200])
```

> Reference: [numpy.where()](https://numpy.org/doc/stable/reference/generated/numpy.where.html)

## Section 8 — Combining DataFrames: Check Your Understanding + Challenges

> Source: `3_More_Pandas/02_Combining_Dataframes.ipynb`  
> Reference: [Pandas merge/join/concat guide](https://pandas.pydata.org/docs/user_guide/merging.html)

### Q1: What is the key difference between `df.join()` and `pd.merge()`?

**Answer:**

- **`df.join()`** aligns on the **index** by default. It is the fastest option when two DataFrames already share the same index (e.g. when you create a derived column as a Series and want to add it back).

- **`pd.merge()`** aligns on **column values** by default. Use it when joining on a shared column (like an ID or date) rather than the index. It is equivalent to a SQL JOIN.

Use `df.join()` when your indices already align. Use `pd.merge()` when matching on a specific column.

### Q2: `pd.concat([df_a, df_b])` returns duplicate index values. What caused this? How to fix it?

**Answer:**

By default, `pd.concat()` with `axis=0` stacks DataFrames vertically and **preserves the original index labels** from each. If both DataFrames have indices that include the same values (e.g. both start from 0), the result will contain duplicate index labels.

Fix: pass `ignore_index=True` to reset the index to a new sequential range:

```python
combined = pd.concat([df_a, df_b], ignore_index=True)
```

> Reference: [pd.concat()](https://pandas.pydata.org/docs/reference/api/pandas.concat.html)

### Q3: An inner merge returns fewer rows than either original DataFrame. Why?

**Answer:**

An **inner join** keeps only rows where the key value appears in **both** DataFrames. Any key present in only one DataFrame is discarded. This is by design — inner means "the intersection".

If you want to keep all rows from one or both DataFrames (filling missing values with `NaN`), use `how='left'`, `how='right'`, or `how='outer'`.

> Reference: [Merge how parameter](https://pandas.pydata.org/docs/user_guide/merging.html#database-style-dataframe-or-named-series-joining-merging)

### Q4: `pd.cut(df['price'], bins=4)`. What does 'equal-width' mean? How is it different from equal-count?

**Answer:**

- **Equal-width bins** (`pd.cut()`): divide the *range* of values into intervals of equal size. For example, if prices range from 0 to 100 and you request 4 bins, you get `(0,25]`, `(25,50]`, `(50,75]`, `(75,100]`. The intervals are the same width, but some bins may contain many more values than others.

- **Equal-count bins** (`pd.qcut()`): divide the values so that each bin contains approximately the *same number of observations*. The interval widths will differ.

```python
pd.cut(df['price'], bins=4)    # equal-width intervals
pd.qcut(df['price'], q=4)      # quartiles — each bin ~= 25% of rows
```

> Reference: [pd.cut()](https://pandas.pydata.org/docs/reference/api/pandas.cut.html) | [pd.qcut()](https://pandas.pydata.org/docs/reference/api/pandas.qcut.html)

### Q5: In a pivot table, some cells contain `NaN`. What does that tell you?

**Answer:**

A `NaN` in a pivot table cell means that **no rows in the original DataFrame match that combination** of index value and column value. There is no data for that particular group, so there is nothing to aggregate.

For example, in a pivot of `quality` vs `acidity_bin`, if no wines of quality 3 fall into the highest acidity bin, that cell is `NaN`.

You can fill these with a meaningful placeholder using `fill_value=0` in `pd.pivot_table()`:

```python
pd.pivot_table(df, values='alcohol', index='quality',
               columns='acidity_bin', fill_value=0)
```

> Reference: [pd.pivot_table()](https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html)

## Section 9 — Datetime: Check Your Understanding

> Source: `3_More_Pandas/03_Datetime.ipynb`  
> Reference: [Python datetime docs](https://docs.python.org/3/library/datetime.html) | [Pandas time series guide](https://pandas.pydata.org/docs/user_guide/timeseries.html)

### Q1: The `timestamp` column has dtype `object`. What does this tell you, and what should you do?

**Answer:**

dtype `object` means the column is stored as **Python strings** (or mixed types). Pandas loaded the date text verbatim from the CSV without interpreting it as dates. You **cannot** do date arithmetic, filtering by date range, or extract `.dt` components until you convert it.

Fix:

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
# Or if the format is known (faster and avoids ambiguity):
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%Y-%m-%d')
```

After conversion, the dtype becomes `datetime64[ns]` and the `.dt` accessor becomes available.

> Reference: [pd.to_datetime()](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html)

### Q2: What is the difference between `strptime()` and `strftime()`? When would you use each?

**Answer:**

A simple mnemonic: **p**arse vs **f**ormat.

- **`strptime(string, format)`** — **parses** a string into a `datetime` object. Use it when you have a date as a string (e.g. `'2021-05-14'`) and need to work with it as a date.
- **`strftime(format)`** — **formats** a `datetime` object into a custom string. Use it when you need to display or export a date in a specific format.

```python
from datetime import datetime
dt = datetime.strptime('14/05/2021', '%d/%m/%Y')  # string → datetime
s  = dt.strftime('%B %d, %Y')                     # datetime → 'May 14, 2021'
```

> Reference: [strftime/strptime reference](https://docs.python.org/3/library/datetime.html#strftime-strptime-behavior)

### Q3: You subtract two datetimes and get `365 days, 3600 seconds`. How many hours is the total duration?

**Answer:**

````python
from datetime import timedelta

# 365 days + 3600 seconds
duration = timedelta(days=365, seconds=3600)

# Convert total duration to hours using timedelta division
hours = duration / timedelta(hours=1)
print(f'{hours} hours')
# 365 * 24 = 8760 hours + 1 hour = 8761 hours
````

### Q4: What does the `.dt` accessor do? What must be true about the column first?

**Answer:**

The **`.dt` accessor** gives you access to **datetime component properties and methods** on an entire Pandas Series at once — things like `.dt.day`, `.dt.month`, `.dt.year`, `.dt.weekday`, `.dt.hour`, etc.

The column **must have `datetime64` dtype** before `.dt` can be used. If the column is still `object` (strings), the `.dt` accessor will raise an `AttributeError`.

```python
# Only works after pd.to_datetime() has been applied:
df['month'] = df['date'].dt.month
df['year']  = df['date'].dt.year
df['dow']   = df['date'].dt.weekday  # 0 = Monday
```

> Reference: [.dt accessor](https://pandas.pydata.org/docs/reference/series.html#datetimelike-properties)

### Q5: Why prefer `df['date'] = pd.to_datetime(df['date'])` over `df.date = ...` in Pandas 3.0?

**Answer:**

In **Pandas 3.0**, attribute-style assignment (`df.date = ...`) is deprecated and may not work correctly due to **Copy-on-Write (CoW)** semantics. CoW means that setting a value on a view (rather than on the original DataFrame) may silently fail or raise a warning instead of modifying the DataFrame in place.

Bracket notation (`df['date'] = ...`) is explicit — it always modifies the DataFrame directly and is the recommended pattern going forward.

```python
# Deprecated in Pandas 3.0:
df.date = pd.to_datetime(df.date)

# Correct pattern:
df['date'] = pd.to_datetime(df['date'])
```

> Reference: [Copy-on-Write guide](https://pandas.pydata.org/docs/user_guide/copy_on_write.html)

````python
# Attach bins to the not-top-acidity subsets from Challenge 3
# (Re-create them here cleanly for self-contained execution)
red_df_b   = red_df.copy()
white_df_b = white_df.copy()

red_df_b['acidity_bin']   = pd.cut(red_df_b['fixed acidity'], bins=5)
white_df_b['acidity_bin'] = pd.cut(white_df_b['fixed acidity'], bins=5)

# Exclude top bin
red_sub   = red_df_b[red_df_b['acidity_bin'] != red_df_b['acidity_bin'].max()]
white_sub = white_df_b[white_df_b['acidity_bin'] != white_df_b['acidity_bin'].max()]

# Pivot: average alcohol by quality (rows) and acidity bin (columns)
print('Red wines — average alcohol by quality and acidity bin:')
print(pd.pivot_table(red_sub, values='alcohol',
                     index='quality', columns='acidity_bin', aggfunc='mean'))

print('\nWhite wines — average alcohol by quality and acidity bin:')
print(pd.pivot_table(white_sub, values='alcohol',
                     index='quality', columns='acidity_bin', aggfunc='mean'))
````
